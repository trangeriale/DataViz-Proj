```python
'''
If you haven't had these packages, please install them via pip (e.g pip install requests).
Please go to pip website (https://pip.pypa.io/en/stable/) if you need any installation instructions!
'''
import requests
from bs4 import BeautifulSoup
```

```python
'''
We now get the tables of GPU names, prices and other specs from wikipedia using BeautifulSoup and requests.

'''

BASE_URL = "https://en.wikipedia.org"

#need a header for Wikipedia to accept our requests
HEADERS = {'User-Agent': 'Mozilla/5.0'}

'''
Get the wikitables of GPU prices. The prices are grouped into tables according to generations of GPUs, so we'd have
multiple tables
'''
def get_main_table():
    #make a request to the page of Nvidia's GPU list
    response = requests.get(BASE_URL +
                            '/wiki/List_of_Nvidia_graphics_processing_units',
                            headers=HEADERS)
    
    #BeautifulSoup will create a complex object tree with all HTML elements presented on the page
    soup = BeautifulSoup(response.content, features="html.parser")

    #many tables on this page because
    tables = soup.find_all('table', {'class': 'wikitable'})

    return tables


GPU_list = get_main_table()


#print(GPU_list)
```
